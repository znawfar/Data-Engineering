**Python** supports JSON through a built-in package called **json**.


```python
import json
```

> **Part 1: Writing JSON to a File**

Writing JSON to a file or Serialization is the process of converting an object into a special format which is suitable for transmitting over the network or storing in file or database.

```python
Manager = {
    'First_Name' : 'Zinat',
    'Last_Name' : 'Nawfar',
    'Education' : 'Master in Social Science',
    'Sallary': 50000,
    'Skills': {
        "Python": "Advanced",
        "SQL": "Advanced",
        "Excel": "Advanced",
        "PowerBI": "Intermediate",
    }
}
```

The JSON library in Python uses the dump() or dumps() function to convert the Python objects into their respective JSON object. 



 **Serialization using dump() function:**


 json.dump() method can be used for writing to **JSON file**. It takes two parameters:
 - dictionary
 - file_pointer


```python
with open('Manager.json', 'w') as f: 
    json.dump(Manager, f)
```

**serialization using dumps() function**


json.dumps() that helps in converting a dictionary to a **JSON** **object**.
 It takes two parameters:
 - dictionary
 - indent

```python
json_object = json.dumps(Manager, indent = 4) 
  
# Writing to sample.json 
with open("sample.json", "w") as outfile: 
    outfile.write(json_object) 

print(json_object)
```

>**Part 2: Reading JSON to a File**

Reading JSON to a File or Deserialization converts the special format returned by the serialization back into a usable object.

```python
with open('sample.json', 'r') as openfile: 

    json_object = json.load(openfile) 
  
print(json_object) 
print(type(json_object)) 
```

>**Module 4: XML file format (Extensible Markup Language)**

**Reading with xml.etree.ElementTree**


```python
import xml.etree.ElementTree as etree
```

```python
# get XMl from the url and wite on the GDrive
url = 'http://www.w3schools.com/xml/plant_catalog.xml'

import requests
resp = requests.get(url)
with open('plant_catalog.xml', 'wb') as myoutput:
   myoutput.write(resp.content)
```

```python
tree = etree.parse("plant_catalog.xml")

root = tree.getroot()
columns = ["COMMON", "BOTANICAL", "ZONE", "LIGHT", "PRICE","AVAILABILITY"]

datatframe = pd.DataFrame(columns = columns)

for node in root: 
    COMMON = node.find("COMMON").text
    BOTANICAL = node.find("BOTANICAL").text 
    ZONE = node.find("ZONE").text   
    LIGHT = node.find("LIGHT").text   
    PRICE = node.find("PRICE").text   
    AVAILABILITY = node.find("AVAILABILITY").text    
    datatframe = datatframe.append(pd.Series([COMMON, BOTANICAL, ZONE, LIGHT, PRICE, AVAILABILITY], index = columns), ignore_index = True)
```
